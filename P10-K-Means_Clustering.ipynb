```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('../Iris.csv')

X = df.drop(columns=['Species']).values
y_true = LabelEncoder().fit_transform(df['Species'])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X_scaled)

print("\n--- CLUSTERING METRICS ---")
print(f"K-Means Inertia (WCSS):          {kmeans.inertia_:.2f}")
print(f"K-Means Silhouette Score:        {silhouette_score(X_scaled, y_kmeans):.4f}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 5))

# 1. Visualize Actual Ground Truth
axes[0].scatter(X_pca[:, 0], X_pca[:, 1], c=y_true, cmap='Set1', s=60, edgecolor='k')
axes[0].set_title('Actual Species (Ground Truth)', fontsize=14)
axes[0].set_xlabel('Principal Component 1')
axes[0].set_ylabel('Principal Component 2')

# 2. Visualize K-Means
axes[1].scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, cmap='Set1', s=60, edgecolor='k')
axes[1].set_title('K-Means Clustering', fontsize=14)
axes[1].set_xlabel('Principal Component 1')

plt.tight_layout()
plt.show()
```
