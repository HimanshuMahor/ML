```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split, KFold
from sklearn.linear_model import LogisticRegression, Lasso, Ridge
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, classification_report, roc_auc_score
```

```python
df = pd.read_csv("Iris.csv")
df.head()
```

```python
print('Number of records:',df.shape[0])
print('Number of columns:',df.shape[1])
```

```python
# filtered_df = df[(df["Species"]=="Iris-setosa")|(df["Species"]=="Iris-versicolor")]LinearRegression
filtered_df = df[df["Species"].isin(["Iris-setosa","Iris-versicolor"])].copy()
```

```python
print('Number of records:',filtered_df.shape[0])
print('Number of columns:',filtered_df.shape[1])
filtered_df.tail()
```

```python
# print("We are taking Setosa as 1 and versicolor as 0")
# subset = filtered_df[["SepalLengthCm" , "SepalWidthCm", "Species"]]
# print(subset)
filtered_df["Species"]= filtered_df["Species"].map({'Iris-setosa':1, 'Iris-versicolor':0})
```

```python
print(filtered_df.head(100))
```

```python
X = filtered_df [["SepalLengthCm", "SepalWidthCm"]].values
Y = filtered_df ["Species"].values
print(X,Y)
```

```python
scalar = StandardScaler()
X_normalized = scalar.fit_transform(X)

k = 5
kf = KFold(n_splits=k, shuffle = True, random_state = 42)

test_accuracy = []
```

```python
for fold, (train_index,test_index) in enumerate(kf.split(X_normalized),1):
    print(f"\n--FOLD {fold}--")
    
    X_train,X_test = X_normalized[train_index], X_normalized[test_index]
    Y_train,Y_test = Y[train_index], Y[test_index]
    
    model = LogisticRegression()
    model.fit(X_train,Y_train)
    
    w1,w2 = model.coef_[0]
    b = model.intercept_[0]
    print(f"Learned Parameters: wi: {w1}, w2 : {w2}, b : {b}")
    
    Y_trainPred = model.predict(X_train)
    Y_testPred = model.predict(X_test)
    
    train_acc = accuracy_score(Y_train, Y_trainPred)
    test_acc = accuracy_score(Y_test, Y_testPred)
    test_accuracy.append(test_acc)
    
    print(f"Train_acc: {train_acc}")
    print(f"Test_acc: {test_acc}")
    
    print("\n --- Classification Report---- ")
    print(classification_report(Y_test, Y_testPred))
    
    y_prob= model.predict_proba(X_test)[:,1] 
    auc_value = roc_auc_score(Y_test, y_prob)
    print(f"AUC value : {auc_value:.4f}")
    print("-" *30)
```

```python
print(f"\nAVERAGE ACCURACY : {np.mean(test_accuracy):.4f}")
```
